```python
object_coloumn=['Team 1','Team2']
trainX=X.copy()
testX=test.copy()
encoder=OrdinalEncoder()
trainX[object_coloumn]=encoder.fit_transform(X[object_coloumn])
testX[object_coloumn]=encoder.transform(test[object_coloumn])

```

```python
import lightgbm as lgb
def LGBM_model():
    lgb_train = lgb.Dataset(X_train, y_train)
    lgb_valid = lgb.Dataset(X_valid, y_valid)
    params={}
    params['learning_rate']=0.01
    params['boosting_type']='goss' #GradientBoostingDecisionTree
    params['objective']='binary' #Binary target feature
    params['metric']='binary_logloss' #metric for binary classification
    params['max_depth']=12
    params['num_leaves']=60
    params['min_data_in_leaf']=20
    #params['extra_trees']=True
#     params['n_jobs']=3
    #params['early_stopping_round']=5
    #params['lambda_l2']=0.01
    num_round = 2450
    bst = lgb.train(params, lgb_train, num_round, valid_sets=[lgb_valid],early_stopping_rounds=100)
    #bst=lgb.cv(params, lgb_train, num_round, nfold=5)
    return bst
```

```python
model=LGBM_model()
```

```python
model.params
```

```python
pred=model.predict(X_valid)
pred1=pred.copy()
pred1
```

```python
pred=pred.round(0)
pred.astype(int)
pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_valid, pred)
print('Confusion matrix\n\n', cm)
print('\nTrue Positives(TP) = ', cm[0,0])
print('\nTrue Negatives(TN) = ', cm[1,1])
print('\nFalse Positives(FP) = ', cm[0,1])
print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
import seaborn as sns
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```
